# Exploratory Data Analysis for TEMP

Type: Temperature sensor\
Unit: °C

**Why temperature helps detect stress**

Stress causes vasoconstriction (blood vessels tighten).
| State   | Skin Temperature   |
| ------- | ------------------ |
| Relaxed | stable             |
| Stress  | slightly decreases |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../Dataset/S2_chest_data.csv")
accept_label = [1, 2, 3, 4]
df = df[df["Label"].isin(accept_label)]

nStr = [1, 3, 4]
def apply_target(label):
    if label == 2:
        return "stressed"
    elif label in nStr:
        return "not stressed"
    else:
        return "unknown"

df["target"] = df["Label"].apply(apply_target)

df.head(2)
```

### EDA start from here

```python
df["Temp"].describe()
```

```python
plt.scatter(range(len(df)), df["Temp"], s=1)
plt.title("Temp signal")
plt.show()
```

```python
str_EDA = df[df["target"] == "stressed"]
nStr_EDA = df[df["target"] == "not stressed"]
```

```python
plt.scatter(range(len(str_EDA)), str_EDA["Temp"], s=1)
plt.title("Temp signal for str_EDA")
plt.show()

plt.scatter(range(len(nStr_EDA)), nStr_EDA["Temp"], s=1)
plt.title("Temp signal for nStr_EDA")
plt.show()
```

stress class temp strongly says that if your body temperature is between 31 - 32 you might be in stress but non stress class is too much randomize and hard to predict

```python
plt.boxplot(str_EDA["Temp"])
plt.title("Temp Distribution for stress class")
plt.show()

plt.boxplot(nStr_EDA["Temp"])
plt.title("Temp Distribution for not stress class")
plt.show()
```
